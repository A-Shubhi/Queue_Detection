Id is generated for the whole video but the cumulative count is for the vehicles inside the box and the results are getting stored in a csv with columns time, frame, class wise count total

```python
import cv2
import numpy as np
from collections import defaultdict
import supervision as sv
from ultralytics import YOLO
import pandas as pd

# Load your custom trained model
model = YOLO('/content/drive/MyDrive/last (1).pt')

# Define the area of interest
area = np.array([(1178, 688), (597, 695), (705, 29), (841, 61)], dtype=np.float32)

# Get class names
class_names = {0: 'car', 1: 'bus', 2: 'motorcycle', 3: 'NMC', 4: 'pedestrian', 5: 'truck', 6: 'LCV', 7: 'Auto'}

# Initialize cumulative class counts
cumulative_class_counts = defaultdict(int)

# Initialize a dictionary to store class-wise detection data
class_detection_data = {class_name: defaultdict(list) for class_name in class_names.values()}

# Open the video capture
cap = cv2.VideoCapture("/content/drive/MyDrive/vid1.mp4")

# Get the frame rate of the video
fps = cap.get(cv2.CAP_PROP_FPS)

# Open a video sink for the output video
video_info = sv.VideoInfo.from_video_path("/content/drive/MyDrive/vid1.mp4")
with sv.VideoSink("/content/drive/MyDrive/res/output_area_count.avi", video_info) as sink:

    frame_number = 0

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.track(frame, classes=[0, 1, 2, 3, 4, 5, 6, 7], persist=True, tracker="bytetrack.yaml")

        # Get the boxes and track IDs
        if results[0].boxes is not None:
            boxes = results[0].boxes.xywh.cpu().numpy()
            track_ids = results[0].boxes.id.int().cpu().tolist() if results[0].boxes.id is not None else []
            classes = results[0].boxes.cls.int().cpu().tolist() if results[0].boxes.cls is not None else []

            # Filter detections based on the area
            for box, track_id, cls in zip(boxes, track_ids, classes):
                x, y, w, h = box
                center = (int(x), int(y))

                # Check if the object center is within the defined area
                if cv2.pointPolygonTest(area, center, False) >= 0:
                    class_name = class_names[cls]
                    cumulative_class_counts[class_name] += 1

                    # Get the current timestamp
                    seconds = frame_number / fps
                    hours, seconds = divmod(seconds, 3600)
                    minutes, seconds = divmod(seconds, 60)
                    timestamp = f"{int(hours):02}:{int(minutes):02}:{int(seconds):02}"

                    # Append the detection data to the class-wise detection dictionary
                    class_detection_data[class_name]["timestamp"].append(timestamp)
                    class_detection_data[class_name]["frame"].append(frame_number)
                    class_detection_data[class_name]["count"].append(cumulative_class_counts[class_name])

            # Annotate the frame with the filtered detections
            annotated_frame = frame.copy()
            for box in boxes:
                x, y, w, h = box
                cv2.rectangle(annotated_frame, (int(x - w / 2), int(y - h / 2)), (int(x + w / 2), int(y + h / 2)), (0, 255, 0), 2)
        else:
            # If no detections, just annotate the frame without any boxes
            annotated_frame = frame

        # Draw the rectangular area on the frame
        cv2.polylines(annotated_frame, [area.astype(np.int32)], isClosed=True, color=(0, 255, 0), thickness=2)

        # Write the cumulative count of objects on each frame for each class
        y_offset = 30
        for class_name, count in cumulative_class_counts.items():
            count_text = f"{class_name}: {count}"
            cv2.putText(annotated_frame, count_text, (10, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            y_offset += 30

        # Write the frame with annotations to the output video
        sink.write_frame(annotated_frame)

        # Increment the frame number
        frame_number += 1

# Release the video capture
cap.release()

# Convert class-wise detection data dictionary to DataFrame
dfs = {}
for class_name, data in class_detection_data.items():
    dfs[class_name] = pd.DataFrame(data)

# Save class-wise detection data to CSV files
for class_name, df in dfs.items():
    df.to_csv(f"/content/drive/MyDrive/res/{class_name}_detections.csv", index=False)

# Merge DataFrames based on timestamp and frame
merged_df = pd.DataFrame()
for class_name, df in dfs.items():
    df["class_name"] = class_name
    if merged_df.empty:
        merged_df = df
    else:
        merged_df = pd.concat([merged_df, df], ignore_index=True)

# Save the merged DataFrame to a CSV file
merged_df.to_csv("/content/drive/MyDrive/res/detections.csv", index=False)
```

```python
# Set up video capture
cap = cv2.VideoCapture("testing/d.mp4")

# Define the line coordinates
START = sv.Point(182, 254)
END = sv.Point(462, 254)


# Store the track history
track_history = defaultdict(lambda: [])

# Create a dictionary to keep track of objects that have crossed the line
crossed_objects = {}

# Open a video sink for the output video
video_info = sv.VideoInfo.from_video_path("testing/d.mp4")
with sv.VideoSink("output_single_line.mp4", video_info) as sink:

    while cap.isOpened():
        success, frame = cap.read()

        if success:
            # Run YOLOv8 tracking on the frame, persisting tracks between frames
            results = model.track(frame, classes=[2, 3, 5, 7], persist=True, save=True, tracker="bytetrack.yaml")

            # Get the boxes and track IDs
            boxes = results[0].boxes.xywh.cpu()
            track_ids = results[0].boxes.id.int().cpu().tolist()

            # Visualize the results on the frame
            annotated_frame = results[0].plot()
            detections = sv.Detections.from_yolov8(results[0])

            # Plot the tracks and count objects crossing the line
            for box, track_id in zip(boxes, track_ids):
                x, y, w, h = box
                track = track_history[track_id]
                track.append((float(x), float(y)))  # x, y center point
                if len(track) > 30:  # retain 30 tracks for 30 frames
                    track.pop(0)

                # Check if the object crosses the line
                if START.x < x < END.x and abs(y - START.y) < 5:  # Assuming objects cross horizontally
                    if track_id not in crossed_objects:
                        crossed_objects[track_id] = True

                    # Annotate the object as it crosses the line
                    cv2.rectangle(annotated_frame, (int(x - w / 2), int(y - h / 2)), (int(x + w / 2), int(y + h / 2)), (0, 255, 0), 2)

            # Draw the line on the frame
            cv2.line(annotated_frame, (START.x, START.y), (END.x, END.y), (0, 255, 0), 2)

            # Write the count of objects on each frame
            count_text = f"Objects crossed: {len(crossed_objects)}"
            cv2.putText(annotated_frame, count_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Write the frame with annotations to the output video
            sink.write_frame(annotated_frame)
        else:
            break

# Release the video capture
cap.release()
```
